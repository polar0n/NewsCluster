```python
import nltk
```

```python
contraction_mapping = {
    "don't": "do not",
    "can't": "cannot",
    "won't": "will not",
    "i'm": "i am",
    "i've": "i have",
    "i'll": "i will",
    "he'll": "he will",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "they're": "they are",
    "they'll": "they will",
    "we're": "we are",
    "we'll": "we will",
    "you're": "you are",
}
```

```python
all_nouns = list()
for filename in os.listdir('paragraphs/'):
    with open(f'paragraphs/{filename}', 'r', encoding='utf-8') as f:
        tokens = nltk.tokenize.TreebankWordTokenizer().tokenize(f.read())
    expanded_tokens = tokens
    for i in range(len(tokens)):
        if tokens[i] in contraction_mapping.keys():
            expanded_tokens[i] = contraction_mapping[tokens[i]]
    nouns = set()
    for (word, pos) in nltk.pos_tag(tokens):
        if pos == 'NN' and word.isalpha() and len(word) > 2:
            insertion = word
            if '.' in insertion:
                insertion.replace('.', '')
            nouns.add(word)
    nouns = list(nouns)
    with open(f'nouns/{filename[:-4]}.json', 'w', encoding='utf-8') as f:
        json.dump(nouns, f)
    all_nouns.append(nouns)
```
