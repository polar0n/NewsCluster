```python
import os
import json
import numpy as np
```

```python
ldir = os.listdir('nouns/')
```

```python
ldir
```

```python
nouns = dict()
```

```python
for filename in ldir:
    with open(f'nouns/{filename}') as f:
        nouns[filename[:-5]] = json.load(f)
```

```python
nouns
```

```python
word_dict = dict()
```

```python
for nouns_list in nouns.values():
    for noun in nouns_list:
        if not word_dict.get(noun):
            word_dict[noun] = 1
        else:
            word_dict[noun] += 1
```

```python
# How many word-articles count dictionary
word_dict
```

```python
word_dict.keys()
```

```python
# Total amount of nouns identified in all the articles
len(word_dict)    
```

```python
word_dict_occ = dict()
for (key, value) in word_dict.items():
    if value >= 20:
        word_dict_occ[key] = value
word_dict = word_dict_occ
```

```python
# Total amount of nouns in at least 20 articles
len(word_dict)
```

Words that can be found in at least 20 articles

```python
words = np.array(list(word_dict.keys()))
```

```python
words
```

```python
np.save('words.npy', words)
```

### Compute word histogram for articles ###

```python
corpus = list()
for filename in os.listdir('paragraphs/'):
    histogram = list()
    with open(f'paragraphs/{filename}', 'r', encoding='utf-8') as file:
        paragraphs = file.read()
        for word in word_dict.keys():
            histogram.append(paragraphs.count(word))
    corpus.append(histogram)
```

```python
with open('corpus.json', 'w', encoding='utf-8') as file:
    json.dump(corpus, file, indent=2)
```
